```python
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.head()
```

```python
df.columns = ['y']
```

```python
df['ds'] = df.index
```

```python
df.head()
```

```python
df.tail()
```

```python
m = Prophet()
```

```python
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=12, freq='MS')
```

```python
future.tail()
```

```python
forecast = m.predict(future)
```

```python
fig = m.plot(forecast);
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

```python
m.plot_components(forecast);
```

```python
m2 = Prophet(seasonality_mode='multiplicative')
```

```python
m2.fit(df)
```

```python
future = m2.make_future_dataframe(periods=12, freq='MS')
```

```python
forecast = m2.predict(future)
```

```python
fig = m2.plot(forecast);
a = add_changepoints_to_plot(fig.gca(), m2, forecast)
```

```python
m2.plot_components(forecast);
```

## Outliers

```python
df.loc['1955-01-01', 'y'] = 600
df.loc['1957-06-01', 'y'] = 1
```

```python
df['y'].plot(figsize=(15, 5));
```

```python
m = Prophet(seasonality_mode='multiplicative')
m.fit(df)
future = m.make_future_dataframe(periods=12, freq='MS')
forecast = m.predict(future)
m.plot(forecast); # wide confidence interval
```

```python
df2 = df[~df.index.isin(['1955-01-01', '1957-06-01'])]
```

```python
df2['y'].plot(figsize=(15, 5));
```

```python
m = Prophet(seasonality_mode='multiplicative')
m.fit(df2)
future = m.make_future_dataframe(periods=12, freq='MS')
forecast = m.predict(future)
m.plot(forecast);
```
